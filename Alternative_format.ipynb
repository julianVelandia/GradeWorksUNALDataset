```python
from datasets import load_dataset, Dataset
import pandas as pd
from huggingface_hub import notebook_login

notebook_login()
```

```python
import json

with open("dataset.json", "r", encoding="utf-8") as f:
    data = json.load(f)
```

```python
rows = []
for record in data.values():
    if "raw_content" not in record:
        print("Missing field: raw_content")
        record["raw_content"] = ""
    rows.append(record)

df = pd.DataFrame(rows)
```

```python
print("Shape:", df.shape)
print("Record:", {k: (v[:500] if k == "raw_content" else v) for k, v in df.iloc[10].to_dict().items()})
```

```python
dataset = Dataset.from_pandas(df)
dataset.push_to_hub("JulianVelandia/unal-repository-dataset-alternative-format")

print("Uploaded successfully.")
```
